# Functional Programming

```python
import random

data = [ (random.randint(1, 1000), random.randint(1, 1000)) for _ in range(100000)]

def gcd(a,b):
    r = a % b
    while r != 0:
        a = b
        b = r
        r = a % b
    return b
```

```python
%%time
res = []
for pair in data:
    n1,n2 = pair[0],pair[1]
    m = gcd(n1,n2)
    res.append((int(n1/m),int(n2/m)))

print(res)
```

```python
def simplify(pair):
    n1,n2 = pair[0],pair[1]
    m = gcd(n1,n2)
    return (int(n1/m),int(n2/m))
```

```python
%%time

res1 = [ simplify(pair) for pair in data ]
print(res1)
```

```python
%%time

print(list(map(simplify,data)))
```
