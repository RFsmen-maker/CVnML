# 读入、显示、导出图片
* 读入、保存图像
* Exp：加载一个图片，显示，按下 S 键保存后退出，或者 ESC 键不保存

```python
# -*- coding: utf-8 -*-

import cv2
from matplotlib import pyplot as plt
import numpy as np

tomatosPath = 'pics/tomatos.jpg'
tomatoWin = 'Pic:tomatos'

tomatosImg_gray = cv2.imread(tomatosPath, 0)
cv2.namedWindow(tomatoWin, cv2.WINDOW_AUTOSIZE)  # WINDOW_NORMAL 可调窗口大小
cv2.imshow(tomatoWin, tomatosImg_gray)  # 直接打开一个窗口显示图片

k = cv2.waitKey(0)  # 键盘绑定函数，单位 ms，0 等待键盘输入，没有输入返回 -1
if k == 27:
    cv2.destroyAllWindows()
elif k == ord('s'):  # s 字符串转 ASCII 对应整数
    cv2.imwrite('testSave.jpg', tomatosImg_gray)
    cv2.destroyWindow(tomatoWin)
else:  # 防止输入其他键时程序挂掉
    cv2.destroyAllWindows()
```

# 设备摄像头使用
* 摄像头捕获：camera = cv2.VideoCapture()
* Exp：获取电脑摄像头，录制倒转视频
* Note：确保安装了版本合适的 ffmpeg, gstreamer

```python
camera = cv2.VideoCapture(0)

# 录制视频
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # 定义 codec
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640,480))  # 创建 VideoWriter 对象

# camera.get(CV_CAP_PROP_)  # 获得视频参数信息，0 到 18，代表视频属性，3 宽 4 高
# camera.set(CV_CAP_PROP_, value)  # 修改视频参数

if camera.isOpened():  # 检查是否成功初始化
    print("camera is opened")
else:
    camera.open()  # 初始化失败要尝试重新打开

while(camera.isOpened()):
    ret, frame = camera.read()  # frame-by-frame，返回布尔值，检查视频文件是否到结尾
    if ret == True:
        frame = cv2.flip(frame, 0)  # 倒转视频
        out.write(frame)
        cv2.imshow('frame', frame)
        # gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break        
    else:
        break

out.release()
camera.release()
cv2.destroyAllWindows()
```

# 图形绘制
* 绘图：line(), rectangle(), circle(), ellipse(), fillPoly(), polylines(), putText()
* 需要输入参数：
|参数|描述|
|----|----|
|img|目标图像|
|color|形状颜色，RGB, (255,0,0) 红色，灰度图只有一个值|
|thickness|线条粗细，-1 闭合，1 默认|
|linetype|线条类型，8 连接、抗锯齿等，默认 8连接，LINE_AA 抗锯齿|

```python
# -*- coding: utf-8 -*-

import cv2
from matplotlib import pyplot as plt
import numpy as np

blackBoard = np.zeros((1000, 1000, 3), np.uint8)

# 红色直线，两端点坐标，颜色，宽度
cv2.line(blackBoard, (0,0), (500,500), (255,0,0), 5)

# 绿色矩形，对角坐标...
cv2.rectangle(blackBoard, (250,250), (750,750), (0,255,0), 3)

# 蓝色实心圆，圆心，半径
cv2.circle(blackBoard, (500,500), 100, (0,0,255), -1)

# 椭圆，圆心，长轴，短轴，椭圆整体逆时针角度，顺时针角度绘制范围（以指定角度起始）
cv2.ellipse(blackBoard, (500,500), (200,50), 0, 0, 90, 255, -1)

# 多边形
polygon = np.array([[[100,100], [200,230], [150,200], [100,220]]], dtype = np.int32)
polyline = np.array([[[0,0], [800,1000], [500,500], [1000,800]]], dtype = np.int32)
# polygon = polygon.reshape((-1, 1, 2))
cv2.fillPoly(blackBoard, polygon, (255,255,0))
cv2.polylines(blackBoard, polyline, True, (255,255,0))  # True 闭合曲线

# 添加文字：内容、位置、字体、线条
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blackBoard, 'OpenCV', (10,500), font, 4, (255,255,255), 2)

plt.imshow(blackBoard)
plt.show()
```

# 鼠标响应
* setMouseCallback()
* Exp：白板画，左键画线，右键画红点，按下 m 切换成：左键画矩形，右键画圆形

```python
# -*- coding: utf-8 -*-

import cv2
from matplotlib import pyplot as plt
import numpy as np

events = [i for i in dir(cv2) if 'EVENT' in i]
print(events)  # 各种鼠标响应信号
```

```python
from math import floor, sqrt, pow

startX, startY = 0, 0
board = np.zeros((1000, 1000, 3), np.uint8)
board.fill(255)
tmpBoard = board.copy()
drawing, mode, rectFlag = False, False, True
windName = "White Board"


def chgsta(x, y):
    global drawing, startX, startY, board, tmpBoard

    if drawing:
        board = tmpBoard.copy()
    else:
        startX, startY = x, y
        tmpBoard = board.copy()
    drawing = not drawing


def draw(event, x, y, flags, param):
    global drawing, mode, rectFlag, board, tmpBoard, startX, startY

    if drawing:  # 画图中
        tmpBoard = board.copy()  # 保留原图
        if mode:  # 模式 2
            if rectFlag:
                cv2.rectangle(tmpBoard, (startX, startY), (x, y), (0, 0, 0), 2)
                if event == cv2.EVENT_FLAG_LBUTTON:
                    chgsta(x, y)
            else:
                cv2.circle(tmpBoard, (startX, startY),
                           int(np.sqrt((y - startY)**2) + (x - startX)**2),
                           (0, 0, 0), 2)
                if event == cv2.EVENT_FLAG_RBUTTON:
                    chgsta(x, y)
                    rectFlag = True
        else:  # 模式 1
            cv2.line(tmpBoard, (startX, startY), (x, y), (0, 0, 0), 2)
            if event == cv2.EVENT_FLAG_LBUTTON:  # 松开画下直线
                chgsta(x, y)
    else:  # 等待画图指令
        if mode:  # 模式 2
            if event == cv2.EVENT_FLAG_LBUTTON:
                chgsta(x, y)
                rectFlag = True
            elif event == cv2.EVENT_FLAG_RBUTTON:
                chgsta(x, y)
                rectFlag = False
        else:  # 模式 1
            if event == cv2.EVENT_FLAG_LBUTTON:
                chgsta(x, y)
            elif event == cv2.EVENT_FLAG_RBUTTON:  # 右键画红点
                cv2.circle(board, (x, y), 5, (0, 0, 255), -1)  # 暂时是 BGR


cv2.namedWindow(windName)
cv2.setMouseCallback(windName, draw)

while True:
    if drawing:
        cv2.imshow(windName, tmpBoard)
    else:
        cv2.imshow(windName, board)
    k = cv2.waitKey(5) & 0xff
    if k == 27:  # ESC 退出
        break
    elif k == ord('m'):  # m 切换模式，左键画矩形
        mode = not mode

cv2.destroyAllWindows()

plt.imshow(cv2.cvtColor(board, cv2.COLOR_BGR2RGB))
plt.show()
```
