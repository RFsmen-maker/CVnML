# numpy, matplotlib, cv2

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

picsPath = '../../../pics/'

def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap = 'gray', vmin = 0, vmax = 255) # 指定显示范围
    else:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()

def gaussian_noise(img, mean, sigma):
    img = img / 255
    noise = np.random.normal(mean, sigma, img.shape)
    gaussian_out = img + noise
    gaussian_out = np.clip(gaussian_out, 0, 1)
    gaussian_out = np.uint8(gaussian_out * 255)
    return gaussian_out
```

* 通道分裂与合并

```python
img = cv.imread(picsPath + 'Ximages/mark.tif')
show(img)
b, g, r = cv.split(img)  # 分裂通道
show(np.hstack([b, g, r]))
show(cv.merge([r, g, b]))  # 合并

# 通道均匀合并成灰色图片
gray = 1/3 * b + 1/3 * g + 1/3 * r  
show(gray)

# 类型转换
gray1, gray2, gray3 = np.uint8(gray), gray.astype(np.uint8), np.float32(gray)
print(gray1.dtype, gray2.dtype, gray3.dtype)
```

* 图像二值化

```python
thresh = 125
# 二值化
gray[gray > thresh] = 255
gray[gray <= thresh] = 0

th1, th2 = 125, 150
# 大于 th1 的都转换成 th2，小于等于 th1 的都转换成 0
left, img_bin = cv.threshold(gray, th1, th2, cv.THRESH_BINARY)
show(np.hstack([gray, img_bin]))
```

* 图像运算：想要相等像素大小图片
    * 相加：混合图像、添加噪声
    * 相减：消除背景、差影法（比较差异、运动跟踪）
    * 相乘：掩膜（mask）
    * 相除：校正设备、比较差异

```python
maize = cv.imread(picsPath + 'FinalAssignment/cim004.jpg')
background = cv.imread(picsPath + 'FinalAssignment/cim008.jpg')
mask = cv.imread(picsPath + 'FinalAssignment/cim014.bmp')

noise = gaussian_noise(maize, 0, 0.8) # 添加高斯噪声的玉米图
left, mask = cv.threshold(mask, 100, 256, cv.THRESH_BINARY)
img_add = maize + background
show(np.hstack([maize, background, mask]))
show(np.hstack([noise, img_add]))
```

```python
img_add1 = cv.add(maize, background) # 一般相加，当超过 255，限制在 255
img_add2 = cv.addWeighted(maize, 0.5, background, 0.5, 20) # 权重相加
img_sub1 = cv.subtract(img_add1, background)
show(np.hstack([img_add1, img_add2, img_sub1]))
```

```python
img_mul1 = maize * mask
img_mul2 = cv.multiply(maize, mask) # 只支持同种数据类型
show(np.hstack([img_mul1, img_mul2]))
```

```python
img_div = cv.divide(maize, noise)
show(img_div)
```

* 像素线性变换

$ s_1 = \beta + \alpha r $

$ s_2 = \alpha + \frac{ln(r + 1)}{\beta} $

$ s_3 = cr^\gamma $

```python
tomatos = cv.imread(picsPath + "Ximages/tomatos.jpg", 0)

b, k = 20, 2
tomato2 = b + k * tomatos.astype(np.int32)
np.clip(tomato2, 0, 255) # 小于 0 的归 0，大于 255 的归于 255
tomato3 = cv.convertScaleAbs(tomatos, alpha = 2, beta = 20) # 取绝对值，并在 255 封顶

tomato4 = 10 + np.log(tomatos.astype(np.float32) + 1) / 0.1

tomato5 = tomatos / 255
tomato5_05 = np.power(tomato5, 0.5) * 255
tomato5_15 = np.power(tomato5, 1.5) * 255

show(np.hstack([tomatos, tomato2, tomato3, tomato4, tomato5_05, tomato5_15]))
```
