Part 1：图像处理——图像分割（Segmentation）
======================================
&emsp;&emsp;图像分割指从图像中提取指定属性，主要基于图像灰度值的**不连续性**和**相似性**，有两类方法：  
* 根据灰度值突变，通过梯度（Gradient）计算出图像的边缘，如 Canny 算法、Sobel 算法等。  
* 根据一组预定义的准则进行整体划分，如阈值处理、区域生长、区域分类、聚合等。  


# 霍夫变换（Hough Transform）
&emsp;&emsp;对于没有明显灰度突变的情况，不能用梯度边缘识别算法识别，如果已知要寻找的物品，可以通过统计像素点的分布信息进行指定形状轮廓的识别。

&emsp;&emsp;以识别直线为例，在得到一个**二值边缘图**后，以 $(x_i,y_i)$ 表示平面上的一个轮廓像素点，已知每个点可以穿过无数条直线，以一定的角度间隔限制穿过这个坐标点的直线数量，在这些直线上作连接坐标系原点的**法线**，每一条法线都可以用一对极坐标参数 $\rho$ 和 $\theta$ 表示，将这些参数点放到以 $\rho$ 和 $\theta$ 作为纵轴与横轴的直角坐标系连接起来即可得到一些列正弦曲线，每一条正弦曲线对应原图中一个轮廓像素点，而这些正弦曲线的**”交点“**即代表原图中存在的一个线段，由于需要进行统计，将第二个直角坐标系划分成多个累加单元，选取累加最大几个作为直线目标。

![pic](pics/hough.png "Hough Transform Principle")

* 参数转换公式：
$$
\rho = x cos\theta + y sin \theta
$$

* 普通的霍夫变换得到的是一条无限长的直线，然而我们需要的是线条，要实现的方式有很多，比如将识别出的直线与检测出的二值轮廓作与运算。

* Matlab：霍夫变换（Hough Transform）识别直线

* Python:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt


circuit = cv2.imread("pics/circuit.tif", cv2.IMREAD_GRAYSCALE)

rows, cols = circuit.shape
roang = 33
M = cv2.getRotationMatrix2D((cols/2, rows/2), roang, 1.0)  # 旋转 45° 不缩放
circuit_rotate = cv2.warpAffine(circuit, M, (rows, cols))
circuit_edges = cv2.Canny(circuit_rotate, 20, 100)

hls = cv2.HoughLinesP(circuit_edges, 1, np.pi/180, 150, 300, 5)  # 1 像素精度，每 ° 间隔
num_line = len(hls)
circuit_lines = cv2.cvtColor(circuit_rotate.copy(), cv2.COLOR_GRAY2RGB)
for i in range(num_line):
    x1, y1, x2, y2 = hls[i][0]
    cv2.circle(circuit_lines, (x1, y1), 5, (255,0,0), -1)
    cv2.line(circuit_lines, (x1, y1), (x2, y2), (0,255,0), 2)
    cv2.circle(circuit_lines, (x2, y2), 5, (0,0,255), -1)
    # print("line :", x1, y1, x2, y2)  # 显示线条坐标

plt.figure(1, (16,32))
plt.subplot(141), plt.imshow(circuit, 'gray'), plt.axis('off'), plt.title("origin")
plt.subplot(142), plt.imshow(circuit_rotate, 'gray'), plt.axis('off'), plt.title("rotate")
plt.subplot(143), plt.imshow(circuit_edges, 'gray'), plt.axis('off'), plt.title("Canny")
plt.subplot(144), plt.imshow(circuit_lines, 'gray'), plt.axis('off'), plt.title("Hough")
plt.show()
```

* 霍夫变换也可以用于识别圆形，已知其平面几何方程：

$$
(x - x_0)^2 + (y - y_0)^2 = r^2
$$

* 一个圆对应三个参数，因此累加器成为三维，每个点可以为一个可能存在的多个圆形和半径投票。


* **总结**：霍夫变换可以识别任意能用参数方程表示出的曲线，需要做的是对曲线的参数空间点投票，但**参数越多**，累加器**维度越高**，对于复杂模型，其**算力成本**不可接受，因此霍夫变换通常**用于简单几何图形**的识别，是最好的技术之一。


* 通过简单的大量线段的识别，再加上局部连续线段判断可以获取图像、视频中比较完整的轮廓信息：
> [线条检测器（Line Segment Detector, LSD）](https://www.ipol.im/pub/art/2012/gjmr-lsd/ "https://www.ipol.im/pub/art/2012/gjmr-lsd/")


# 大律分割（Otsu's Image Segmentation）
&emsp;&emsp;依据直方图直接对图像进行阈值分割。

![pic](pics/otsu.png "Otsu's Segmentation")

* 需要注意其无法直接对噪声图像直接处理，阈值分割前先进行滤波即可。

![pic](pics/otsu_noise.png "Otsu's Segmentation on noised")

* 显然我们需要自动地选取这个阈值，而不是手动，考虑方差最小化。

$$
\sigma_w^2(t) = q_1(t)\sigma_1^2(t) + q_2(t)\sigma_2^2(t)
\\
q_1(t) = \sum_{i=1}^tP(i) \quad q_2(t) = \sum_{i=t+1}^IP(i)
\\
\mu(t) = \sum\frac{iP(i)}{q(t)}
\\
\sigma^2(t) = \sum[i-\mu(t)]^2\frac{P(i)}{q(t)}
\\
\sigma^2 = \sigma_w^2(t) + q_1(t)[1 - q_1(t)][\mu_1(t) - \mu_2(t)]^2
$$

&emsp;&emsp;$\sigma^2$ —— 常数值，整体图像的方差，不会改变

* 如果选到的阈值较大，将会把 $\sigma^2$ 放大。


* 要使得 $\sigma_w^2(t)$ 最小，也可以让 $q_1(t)[1 - q_1(t)][\mu_1(t) - \mu_2(t)]^2$ 最大  


* 找到阈值的方法之一：遍历所有灰度值，选出最小方差对应阈值。

* Matlab：大律算法识别硬币

* Python:

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


coins = cv2.imread("pics/coins.png", cv2.IMREAD_GRAYSCALE)

ret, coins_otsu = cv2.threshold(coins, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)  # 大律分割

circles = cv2.HoughCircles(coins_otsu, cv2.HOUGH_GRADIENT, 1, 50, 50, 30, 10, 20, 30)

coins_cir = cv2.cvtColor(coins.copy(), cv2.COLOR_GRAY2RGB)
circles = np.uint16(np.around(circles))

print("(x y radius)\n", circles)
for i in circles[0,:]:
    cv2.circle(coins_cir, (i[0], i[1]), i[2], (0, 255, 0), 3)

plt.hist(coins.ravel(), 256, [0,256]), plt.title("histogram"), plt.show()
plt.subplot(131), plt.imshow(coins, 'gray'), plt.axis('off'), plt.title("coins")
plt.subplot(132), plt.imshow(coins_otsu, 'gray'), plt.axis('off'), plt.title("otsu")
plt.subplot(133), plt.imshow(coins_cir, 'gray'), plt.axis('off'), plt.title("find")

plt.show()
```

# 交互式图像分割（Interactive Segmentation）
&emsp;&emsp;设计算法自动地进行图像分割非常困难，机器无法按照人的生活经验分辨事物，因此考虑加入用户，以用户最低限度的协助完成图像分割任务。

* 仅仅通过颜色划分还不足以完全分割图像，因为前景中可能会出现背景像素点，可以通过计算距离判断这些像素点。

## 算法步骤：

* 特征分布估计：用户在前景和背景区域作不同的**涂鸦**，统计计算各种像素值为前景或背景的概率。    

![pic](pics/probability.png "statistic")

$$
P_F(x) = \frac{P_r(\vec{c_x}|F)}{P_r(\vec{c_x}|F) + P_r(\vec{c_x}|B)}  
$$

* 权重距离转换：遍历图像中所有像素点，计算像素点分别到达两种涂鸦的**最短路径**（通过 Dijkstra 算法），得到到达背景距离 $D_B(x)$ 和前景距离 $D_F(x)$。

![pic](pics/weightedDis.png "distance")

$$
d(s_1, s_2) := \min_{C_{s_1,s_2}} \, \int_{C_{s_1,s_2}}Wds
\\
W := |\nabla{P_F(x)} \cdot \vec{C'}_{s_1,s_2}(x)|
\\
D_l(x) := \min_{s \in \Omega_l} d(s,x), \quad l \in {F,B}
$$

* 精准修正：由于第一次又用户涂鸦计算距离并分割出的图像边缘较粗糙，由算法对已有的边缘分割向内和向外缩小和扩大一段距离，再一次重复前两个处理过程，得到更精细的边缘。

![pic](pics/refine.png "refine algo")

# 图割分割（Graph Cuts）
&emsp;&emsp;借助图论的工具，将一幅图划分为前景和背景。将待分割的图像表示为一个加权无向图，图中每一个节点即一个像素，边由每对节点之间连接形成，每条边的权重 $w(i,j)$ 是节点 $i$ 和 $j$ 之间的相似性函数，以此将整个图的所有节点划分成不同的子集 $V_1,V_2,...,V_K$，子集内的节点相似性高。

![pic](pics/graphCut.png "Graph Cuts")

* 进一步为了判断各个子集为前景或背景，分别加入**源（source）终端节点**和**汇聚（sink）终端节点**连接到各个像素点，其为前景或背景的概率即连接边 **$t$ 链的权重**。

![pic](pics/graphCut_imp.png "divide fore or back")

## Munford-Shah 分割
&emsp;&emsp;一种将性能和效果间折衷公式化的算法，希望简化图像，但同时与原图差别小。通过**约束**可以识别得到的**边缘数量**与**均方差（Mean Square Error）统计**实现：如果有过多边缘，付出更高性能代价，因此使其尽可能小，如果得到图像过于模糊（识别边缘少），与原图像偏差大（均方差过大）。

![pic](pics/Munford_shah.png "Munford shah")

# 视频分割 AE 工具 Roto Brush 背后的算法

* 工具需求：精确、稳定、实时交互且简便、高效耗时少  
* 困难：像素颜色重叠、边缘模糊、动态背景  


## 算法步骤：

* **局部分类（Localized Classifier）**：从一个给定地帧开始，假设当前已经完成了图像分割，以一定的间隔距离以边缘像素点为中心取局部窗口，在这个窗口内，通过原有边缘得到先前形状（Shape Prior），通过像素颜色划分得到颜色分布（Color Models）。

![pic](pics/rotoBrush_cla.png "local window")

* **多帧传播（Multi-frame Propagation）**：先验假设帧与帧之间不会相差太大，上一帧取到的信息作为**训练（training）**，第二帧作为**测试（testing）**继承从上一帧取得的局部窗口，通过高斯模糊使得两帧的形状分割相似，再根据颜色和形状的对比选择识别策略：如果颜色区别明显，让颜色作为置信映射；如果颜色区别不明显，将先前得到的边缘带缩小并作为置信映射。

![pic](pics/rotoBrush_pro.png "propagation")
![pic](pics/rotoBrush_com.png "combination")

* **局部修正（Local Correction）**：有时候算法不一定可靠，我们需要手工修复，以防误差传递，对于错误帧，加入用户手动修复，让修复的局部向后传播一定的帧数。

![pic](pics/rotoBrush_cor.png "local correction")

* **后处理（Post-processing）**：如将背景模糊并保留锐化的前景视频、对视频中追踪的指定事物调光或增加特效。


Part 2：机器学习——梯度下降（Gradient Descent）
=====================================

# 计算原理

* 线性回归中我们需要找到损失最小的对应参数集合：

$$
\theta^* = arg\,\min_{(\theta)}L(\theta)
$$


* 设线性回归中参数为 $\theta$ 且只有两个参数为 $\{\theta_1,\theta_2\}$，增加参数上标为迭代次数。


* 随机选取一个初始值：

$$
\theta^0 = 
\begin{bmatrix}
\theta_1^0
\\
\theta_2^0
\end{bmatrix}
$$

* 通过计算偏微分不断修正参数值：

$$
\begin{bmatrix}
\theta_1^{i+1}
\\
\theta_2^{i+1}
\end{bmatrix}
 = 
\begin{bmatrix}
\theta_1^{i}
\\
\theta_2^{i}
\end{bmatrix}
 - \eta
\begin{bmatrix}
\partial{L}(\theta_1^i)/\partial{\theta_1}
\\
\partial{L}(\theta_2^i)/\partial{\theta_2}
\end{bmatrix}
$$

* 简化成：

$$
\theta^{i+1} = \theta^i - \eta\nabla{L}(\theta^i)
$$

* 对于只有两个参数的情况，可以认为 Gradient 的方向即 Loss 函数等高线的法线方向，但对于更多参数的情况无法指明，转而以参数更新迭代为横坐标，以损失值为纵坐标，可以绘制任何参数个数的情况。


# 优化策略

## 学习率（Learning Rate）
&emsp;&emsp;损失函数计算参数 $\eta$ 即学习率，影响梯度下降的幅度，学习率越大，Loss 变化越快，但我们要找到的是 Loss 最小时的参数组合，因此要避免幅度过大完全跨过极小值的情况，必须合理调节学习率，同时，同一个学习率可能不适用于所有参数，可以每个参数对应一个学习率。

![pic](pics/loss_visualize.png "visualize loss")

* 要使得学习率能够自动调节，可以有不同策略。
* 最常用的一种：一开始较大，而后逐渐减小。

$$
\eta^t = \eta / \sqrt{t + 1}
$$

* 原始方法 Vanilla：

$$
w^{t+1} \leftarrow w^t - \eta^tg^t
$$

* 优化方法 Adagrad：考虑突变值的反差程度，让每个参数的学习率除以之前所有微分值得均方根，突变越大，均方根越大，使得 Loss 变化反而变缓。

* 更新参数跟微分值成正比，是在没有考虑多个参数得条件下才成立，然而，考虑多个参数时需要偏微分。

* 更新参数的最佳步幅可以通过二次微分计算得到，Adagrad 除以均方根本质上是同时考虑二次微分，同时简便了运算（泰勒展开）

$$
w^{t+1} \leftarrow w^t - \frac{\eta^t}{\sigma^t}g^t
\Rightarrow
w^{t+1} \leftarrow w^t - \frac{\eta^t}{\sqrt{\sum^t_{i=0}(g^i)^2}}g^t
$$


* 更快的方法 Stochastic：一次只考虑一个参数，一次只更新这一个参数

## 特征缩放（Feature Scaling）
&emsp;&emsp;不同特征的值数据幅度大小不同时，会使训练出的权重（参数）幅度也不同，使得梯度下降时需要不同的学习率进行回归，让不同特征缩放到同个幅度范围可以让梯度下降计算更简单。

![pic](pics/feature_scaling.png "feature_scaling")

* 常见归一化：让均值为 0，方差为 1

$$
x^r_i \leftarrow \frac{x^r_i - m_i}{\sigma_i}
$$


# 梯度下降数学原理

* 一元泰勒展开式（Taylor Series）：

$$
\begin{aligned}
h(x) & = \sum_{k=0}^{\infty} \frac{h^(k)(x_0)}{k!} (x - x_0)^k
\\
& = h(x_0) + h'(x_0)(x - x_0) + \frac{h''(x_0)}{2!}(x - x_0)^2 + ...
\end{aligned}
$$

* 二元泰勒展开式：

$$
\begin{aligned}
h(x，y) = h(x_0,y_0) + \frac{\partial{h}(x_0,y_0)}{\partial x} (x - x_0) + \frac{\partial{h}(x_0,y_0)}{\partial y} (y - y_0) + G_1(x-x_0)^2 + G_2(y - y_0)^2 + ...
\end{aligned}
$$

* 对于两个参数的线性回归可以二元泰勒展开，精确到一次：假设当前两个 $\theta$ 参数为 $(a,b)$

$$
L(\theta) \approx L(a,b) + \frac{\partial L(a,b)}{\partial \theta_1}(\theta_1 - a) + \frac{\partial L(a,b)}{\partial \theta_2}(\theta_2 - b)
\\
s = L(a,b) ,\quad u = \frac{\partial{L}(a,b)}{\partial \theta_1} ,\quad v = \frac{\partial{L}(a,b)}{\partial \theta_2}
\\
L(\theta) \approx s + u(\theta_1 - a) + v(\theta_2 - b)
$$

* 要找到邻域内的最小 Loss，即要找到平面内圆形范围内的最小值，有：

![pic](pics/gradient_math.png "math")

$$
(\theta_1 - a)^2 + (\theta_2 - b)^2 \le r^2
\\
\Delta \theta_1 = \theta_1 - a ,\quad \Delta \theta_2 = \theta_2 - b
$$

* 显然这个值是 $(u,v)$(Gradient) 相反方向且半径即为 $r$ 的向量：

$$
\begin{bmatrix}
\Delta \theta_1
\\
\Delta \theta_2
\end{bmatrix}
 = 
-\eta
\begin{bmatrix}
u
\\
v
\end{bmatrix}
\Rightarrow
\begin{bmatrix}
\theta_1
\\
\theta_2
\end{bmatrix}
=
\begin{bmatrix}
a
\\
b
\end{bmatrix}
-\eta
\begin{bmatrix}
u
\\
v
\end{bmatrix}
 = 
\begin{bmatrix}
a
\\
b
\end{bmatrix}
-\eta
\begin{bmatrix}
\frac{\partial C(a,b)}{\partial \theta_1}
\\
\frac{\partial C(a,b)}{\partial \theta_2}
\end{bmatrix}
$$

* 以上推理式子成立的前提是泰勒展开的条件，即要使得邻域足够小才能足够精确，其中的学习率 $\eta$ 与邻域大小成正比，因此需要足够小才能保证上述各式成立。


* 找到最小值的方法不只有 Gradient Descent，比如牛顿法，而泰勒还可以展开到更高次项，这些方法的计算成本较高而无法接受，因此还是以 Gradient Descent 为主流。

* 计算出的微分值非常小，却不一定是局部最小。

# Pytorch 框架实现简易线性回归

```python
import numpy as np
import torch
import torch.utils.data as data
import torch.optim as optim
import matplotlib.pyplot as plt
from torch import nn
from torch.nn import init

'''  线性回归构造
class LinearNet(nn.Module):
    def __init__(self, n_feature):
        super(LinearNet, self).__init__()
        self.liner = nn.Linear(n_feature, 1)

    def forward(self, x):
        y = self.linear(x)
        return y
'''

torch.manual_seed(1)
torch.set_default_tensor_type('torch.FloatTensor')
```

* 定义原函数、生成训练数据集并添加噪声

```python
# 原函数
true_w = [2, -3.4]
true_b = 4.2

# 生成训练数据、标记数据并添加噪声
num_inputs = 2  # 特征数
num_examples = 1000
features = torch.tensor(np.random.normal(0, 1, (num_examples, num_inputs)), dtype=torch.float)
labels = true_w[0] * features[:, 0] + true_w[1] * features[:, 1] + true_b
labels += torch.tensor(np.random.normal(0, 0.01, size=labels.size()), dtype=torch.float)
dataset = data.TensorDataset(features, labels)
batch_size = 10
data_iter = data.DataLoader(  # 装载训练数据
    dataset=dataset,  # torch TensorDataset format
    batch_size=batch_size,  # mini batch size
    shuffle=True,  # 打乱数据 (打乱比较好)
    # num_workers=2  # 多线程来读数据，存在运行错误，关闭
)
```

* 数据分布散点图

```python
w1 = []
w2 = []
list_features = features.tolist()
num_features = len(list_features)

for i in range(num_features):
    w1.append(list_features[i][0])
    w2.append(list_features[i][1])

plt.figure(1, (15,30))
plt.subplot(311), plt.scatter(w1, w2, s=1), plt.xlabel("w1"), plt.ylabel("w2")
plt.subplot(312), plt.scatter(w1, labels.tolist(), s = 1), plt.xlabel("w1"), plt.ylabel("y")
plt.subplot(313), plt.scatter(w2, labels.tolist(), s = 1), plt.xlabel("w2"), plt.ylabel("y")
plt.show()
```

* 模型建立

```python
# net = LinearNet(num_inputs)
net = nn.Sequential(
    nn.Linear(num_inputs, 1)
    # 可以加入其他层
    )
layer_num = len(net)
for i in range(layer_num):
    init.normal_(net[i].weight, mean=0.0, std=0.01)  # 权重按照正态分布
    init.constant_(net[i].bias, val=0.0)  # 偏差默认为 0
```

* 损失函数与优化函数定义

```python
loss = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.03)
```

* 训练模型并与原型比较

```python
num_epochs = 3  # 迭代次数
for epoch in range(1, num_epochs+1):
    ls = None
    for X, y in data_iter:
        output = net(X)
        ls = loss(output, y.view(-1, 1))
        optimizer.zero_grad()
        ls.backward()
        optimizer.step()
    print("epoch %d, loss %f" % (epoch, ls.item()))
    dense = net[0]
    print("weight :", true_w, ':', dense.weight.data.tolist()[0], '\n', "bias :", true_b, ':', dense.bias.data.item(), '\n')
```
