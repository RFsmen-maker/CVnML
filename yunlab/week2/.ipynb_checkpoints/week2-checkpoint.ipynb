# 图像修复（复原）
&emsp;&emsp;以某种预定方式改进图像，与图像增强不同，图像复原利用**退化（Degradation）**现象的先验知识，建模退化和加性噪声过程并应用逆过程对退化图像 $g(x,y)$ 进行修复，得到原图像的一个估计 $\hat{f}(x,y)$。

* 对修复过程进行建模：


## 主要六种噪声类型：
![pic](pics/noise_type.png "noise types")

* Gaussian:  
$$
\begin{equation}
p(z) = \frac{1}{\sqrt{2\pi}\sigma}e^{-(z-\bar{z})^2/2\sigma^2}, -\infty < z < \infty
\end{equation}
\tag{1}
$$  
  
  
* Reyleigh:  
$$
\begin{equation}
p(z) = 
\left\{
\begin{array}{}
\frac{2}{b} (z - a)e^{-(z - a)^2 \over b}\quad &z \ge a
\\
0\quad &z \le a
\end{array}
\right. 
\\
\bar{z} = a + \sqrt{\pi b \over 4}, \quad
\sigma^2 = \frac{b(4 - \pi)}{4}
\end{equation}
\tag{2}
$$  
  
  
* Exponential:  
$$
\begin{equation}
p(z) = 
\left\{
\begin{array}{}
ae^{-az} \quad &z \ge 0
\\
0 \quad &z \lt 0
\end{array}
\right. 
\\
\bar{z} = a + \sqrt{\pi b \over 4}, \quad
\sigma^2 = \frac{b(4 - \pi)}{4}
\end{equation}
\tag{3}
$$  
  
  
* Uniform:  
$$
\begin{equation}
p(z)  =
\left\{
\begin{array}{}
\frac{1}{b - a} \quad & a \le z \le b
\\
0 \quad & other
\end{array}
\right.
\\
\bar{z} = \frac{a+b}{2}, \quad
\sigma^2 = 0
\end{equation}
\tag{4}
$$

* MATLAB：加噪

* Python:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# 添加高斯噪声
def gaussian_noise(src, mean=0, sigma=0):
    '''
    按照正态分布，生成整个图片各个点的随机偏移值，叠加到原图，均值一般为 0
    '''
    dst = src.copy()
    noise = np.random.normal(mean, sigma, src.shape)  # 根据均值和标准差生成噪声图
    # print("noise =", noise.astype(np.int8))
    dst = src + noise.astype(np.int8)  # 直接将噪声叠加到原图
    dst = np.clip(dst, 0, 255)
    return dst


# 添加椒盐噪声
def saltnpepper_noise(src, prop):
    dst = src.copy()
    noise_num = int(np.ceil(src.size * prop))
    # print("salt & pepper noise point amount:", noise_num, " /", src.size)
    for i in range(noise_num):
        x = np.random.randint(0, src.shape[0] - 1)
        y = np.random.randint(0, src.shape[1] - 1)
        if np.random.randint(0, 2):
            dst[x, y] = 0
        else:
            dst[x, y] = 255
    return dst

```

```python
img_gray = cv2.imread('pics/saturn.png', cv2.IMREAD_GRAYSCALE)
img_gaussian25 = gaussian_noise(img_gray, 0, 25)
img_snp002 = saltnpepper_noise(img_gray, 0.02)
img_snp02 = saltnpepper_noise(img_gray, 0.2)

cv2.imwrite("pics/gray.png", img_gray)
cv2.imwrite("pics/gaussian_noise25.png", img_gaussian25)
cv2.imwrite("pics/saltnpepper002.png", img_snp002)
cv2.imwrite("pics/saltnpepper02.png", img_snp02)
```

## 维纳滤波 Winener Filtering:
&emsp;&emsp;运动模糊主要通过逆滤波和维纳（最小方差）滤波处理。逆滤波对加性噪声敏感，无法直接使用，而维纳滤波的能有更好的表现。
* ref: <http://en.wikipedia.org/wiki/Wiener_filter>
$$
e^2 = E[(f(x,y) - \hat{f}(x,y))^2],\quad \hat{F}(u, v) = \frac{H^*(u, v)}{H^2(u,v) + \frac{S_\eta}{S_f}}G(u,v) ,\quad \frac{S_\eta}{S_f} = K
$$  
&emsp;&emsp;K —— 常数，代替难以评估的 $S_\eta$ 和 $S_f$  

* Matlab：加噪与去噪

* Python:

```python
import cv2
import numpy as np
from numpy import fft
import matplotlib.pyplot as plt
import math


# 仿真运动模糊
def motion_process(image_size, motion_angle):
    PSF = np.zeros(image_size)
    center_pos = (image_size[0] - 1) / 2
    
    slope_tan = math.tan(motion_angle * math.pi / 180.0)
    slope_cot = 1 / slope_tan
    if slope_tan <= 1:
        for i in range(15):
            offset = round(i * slope_tan)
            PSF[int(center_pos + offset), int(center_pos - offset)] = 1
    else:
        for i in range(15):
            offset = round(i * slope_cot)
            PSF[int(center_pos - offset), int(center_pos + offset)] = 1
    return PSF / PSF.sum()  # 归一化


def make_blurred(src, PSF, eps):
    input_fft = fft.fft2(src)  # 二维数组傅里叶变换
    PSF_fft = fft.fft2(PSF) + eps
    blurred = np.abs(fft.fftshift(fft.ifft2(input_fft * PSF_fft)))
    return blurred


# 添加运动模糊
def motion_blur(src, length=12, angle=45):
    mask = cv2.getRotationMatrix2D((length / 2, length / 2), angle, 1)
    kernel = np.diag(np.ones(length))
    kernel = cv2.warpAffine(kernel, mask, (length, length)) / length

    blurred = cv2.filter2D(np.array(src), -1, kernel)
    cv2.normalize(blurred, blurred, 0, 255, cv2.NORM_MINMAX)
    blurred = np.array(blurred, dtype=np.uint8)
    return blurred


# 逆滤波
def inverse_filter(src, PSF, eps):
    src_fft = fft.fft2(src)
    PSF_fft = fft.fft2(PSF) + eps
    result = np.abs(fft.fftshift(fft.ifft2(src_fft / PSF_fft)))
    return result


# 维纳滤波
def wiener_filter(src, PSF, eps, K=0.01):
    src_fft = fft.fft2(src)
    PSF_fft = fft.fft2(PSF) + eps
    PSF_fft_1 = np.conj(PSF_fft) / (np.abs(PSF_fft) ** 2 + K)
    result = np.abs(fft.fftshift(fft.ifft2(src_fft * PSF_fft_1)))
    return result
```

```python
img_gray = cv2.imread('pics/saturn.png', cv2.IMREAD_GRAYSCALE)
PSF = motion_process(img_gray.shape, 60)
img_gaussian0005 = gaussian_noise(img_gray, 0, 0.005 * (255**2))
img_wiener = wiener_filter(img_gaussian0005, PSF, 1e-3)
img_blurred = cv2.blur(img_gray, (10, 10), borderType = cv2.BORDER_REPLICATE)
img_motionblur = motion_blur(img_gray, 10, 90)

cv2.imwrite("pics/img_gaussian0005.png", img_gaussian0005)
cv2.imwrite("pics/img_wiener.png", img_wiener)
cv2.imwrite("pics/img_blurred.png", img_blurred)
cv2.imwrite("pics/img_motionblur.png", img_motionblur)
```
