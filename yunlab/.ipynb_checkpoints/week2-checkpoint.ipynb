* Gaussian:  
$$
\begin{equation}
p(z) = \frac{1}{\sqrt{2\pi}\sigma}e^{-(z-\bar{z})^2/2\sigma^2}, -\infty < z < \infty
\end{equation}
$$  
  
  
* Reyleigh:  
$$
\begin{equation}
p(z) = 
\left\{
\begin{array}{}
\frac{2}{b} (z - a)e^{-(z - a)^2 \over b}\quad &z \ge a
\\
0\quad &z \le a
\end{array}
\right. 
\\
\bar{z} = a + \sqrt{\pi b \over 4}, \quad
\sigma^2 = \frac{b(4 - \pi)}{4}
\end{equation}
$$  
  
  
* Exponential:  
$$
\begin{equation}
p(z) = 
\left\{
\begin{array}{}
ae^{-az} \quad &z \ge 0
\\
0 \quad &z \lt 0
\end{array}
\right. 
\\
\bar{z} = a + \sqrt{\pi b \over 4}, \quad
\sigma^2 = \frac{b(4 - \pi)}{4}
\end{equation}
$$  
  
  
* Uniform:  
$$
\begin{equation}
p(z)  =
\left\{
\begin{array}{}
\frac{1}{b - a} \quad & a \le z \le b
\\
0 \quad & other
\end{array}
\right.
\\
\bar{z} = \frac{a+b}{2}, \quad
\sigma^2 = 0
\end{equation}
$$
